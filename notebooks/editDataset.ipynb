```python
import nltk
nltk.download('stopwords')
nltk.download('punkt')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import string
```

```python
import requests
stopwords_list = requests.get("https://gist.githubusercontent.com/rg089/35e00abf8941d72d419224cfd5b5925d/raw/12d899b70156fd0041fa9778d657330b024b959c/stopwords.txt").content
stopwords = set(stopwords_list.decode().splitlines())
print(stopwords)
   
```

```python
data.head()
```

```python
def remove_punction_and_stopwords(msg):
   word_tokens = word_tokenize(msg)
   filtered_words = [w for w in msg if w not in word_tokens and w not in string.punctuation]
   new_sentence = ''.join(filtered_words)
   return new_sentence



data['Product']=data['Product'].apply(remove_punction_and_stopwords)
data['Issue']=data['Issue'].apply(remove_punction_and_stopwords)
data['Company']=data['Company'].apply(remove_punction_and_stopwords)
```

```python
def nlkt(val):
    val=repr(val)
    clean_txt = [word for word in val.split() if word.lower() not in stopwords.words('english')]
    nopunc = [char for char in str(clean_txt) if char not in string.punctuation]
    nonum = [char for char in nopunc if not char.isdigit()]
    words_string = ''.join(nonum)
    return words_string


data['Anylize_Text']= data['Product'].astype(str) + ' ' + data['Issue'].astype(str) + ' ' + data['Company'].astype(str)

finallist =[]

for eachlist in data['Anylize_Text']:
    z = nlkt(eachlist)
    finallist.append(z)
```
